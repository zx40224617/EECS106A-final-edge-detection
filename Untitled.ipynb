```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_backboard(image_path):
    
    image = cv2.imread(image_path)
    original = image.copy() 
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.show()
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 3)
    plt.imshow(blurred, cmap = "grey")
    plt.show()
       
    edges = cv2.Canny(blurred, threshold1=100, threshold2=150)
    plt.imshow(edges)
    plt.show()
    
    
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    #print out all contours using a scatter plot for debug
    '''
    for i in range (len(contours)):
        test = contours[i]

        contour_array = np.array(test)
        print(contour_array.shape)
        #print(contour_array)
        x = contour_array[:, :, 0]
        y = contour_array[:, :, 1]
        plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB)) 
        plt.scatter(x, y, color='blue', s=3)  # Plot the points
        plt.title('Detected Edges')
        plt.axis('off')
        plt.show()
        #print(x)
    '''
    
    #cv2.drawContours(original, contours, -1, (255, 0, 0), 3)
    #print(contours)
    #plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    #plt.show()
    
    
    
    for contour in contours:
        # Approximate the contour to get the corner points
        epsilon = 0.02 * cv2.arcLength(contour, True)
        approx_contour = cv2.approxPolyDP(contour, epsilon, True)

        # Check if the contour has less than 10 points (likely a rectangle)
        if len(approx_contour) <= 10:
            #visualize the points
            '''
            print("only 4 points!")
            print(epsilon)
            print(approx_contour)
            x = approx_contour[:, :, 0]
            y = approx_contour[:, :, 1] 
            plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB)) 
            plt.scatter(x, y, color='blue', s=3)  # Plot the points
            plt.title('Detected Edges')
            plt.axis('off')
            plt.show()
            '''
            # Compute the bounding box and draw the rectangle using the approximate contour
            x, y, w, h = cv2.boundingRect(approx_contour)
    
            
            ratio = float(w) / h

            # Filter based on aspect ratio and size 
            if 1.2 < ratio < 3 and w > 30 and h > 30:
                print("rectangle found!")
                #midpoint of the backboard
                midpoint = np.array([(x + w / 2), (y + h / 2)])
                cv2.drawContours(original, [approx_contour], -1, (255, 0, 0), thickness=2)
                


    plt.figure(figsize=(10, 5))
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.scatter(midpoint[0], midpoint[1], color='blue', s=3)
    plt.title("Detected Backboard Rectangle")
    
    plt.axis("off")
    plt.show()

    
detect_backboard('backboard1.jpg')
```
